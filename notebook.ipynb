```
from google.colab import files

uploaded = files.upload()
```

```
import pandas as pd
import numpy as np

# Only one file upload is accepted.
nfiles = len(uploaded.keys())
if nfiles > 1:
    raise NameError(f"only one input file will be accepted, got {nfiles}")

name = ""
for k in uploaded.keys():
    name = k
if name == "":
    raise NameError("no uploaded filename found")

df = pd.read_csv(name)
df.columns = ["ctx", "algorithm", "size", "alloc", "free", "exec"]
```

```
# Dataframe preparation.

dfg = df.groupby(["size", "ctx", "algorithm"])
dfg.agg([np.mean, np.std])
```
